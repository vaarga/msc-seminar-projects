```python
import pandas as pd
import matplotlib.pyplot as plt
```

1. Create a file with the following structure (type, engine, fuel, year, mileage, price), regarding the
data for several SUV cars. Find the minimum and maximum price and the characteristics of the
corresponding cars.

```python
# Create the file programmatically
COLUMN_NAMES = ['type', 'engine', 'fuel', 'year', 'mileage', 'price']
SUV_CARS = [
    ['Ford Expedition', 'V6', 'Gasoline', 2020, 25000, 35000],
    ['Honda CR-V', '4-cylinder', 'Hybrid', 2021, 18000, 42000],
    ['Jeep Wrangler', 'V8', 'Diesel', 2019, 30000, 28000],
    ['Chevrolet Tahoe', '6-cylinder', 'Gasoline', 2022, 15000, 38000],
    ['Tesla Model X', '4-cylinder', 'Electric', 2023, 1000, 55000],
    ['Ford Expedition', 'V6', 'Gasoline', 2020, 22000, 36000],
    ['Honda CR-V', '4-cylinder', 'Hybrid', 2022, 12000, 40000],
    ['Jeep Wrangler', 'V8', 'Diesel', 2021, 28000, 32000],
    ['Chevrolet Tahoe', '6-cylinder', 'Gasoline', 2023, 8000, 42000],
    ['Tesla Model X', '4-cylinder', 'Electric', 2020, 500, 60000]
]
FILE_NAME = 'suv_cars.csv';

suv_cars_df = pd.DataFrame(SUV_CARS, columns=COLUMN_NAMES)

suv_cars_df.to_csv(FILE_NAME, index=False)
```

```python
suv_df = pd.read_csv(FILE_NAME)

suv_df
```

```python
car_with_maximum_price = suv_df[suv_df['price'] == suv_df['price'].max()]

car_with_maximum_price
```

```python
car_with_minimum_price = suv_df[suv_df['price'] == suv_df['price'].min()]

car_with_minimum_price
```

2. Create a plot that shows on one axis the mileage and on the other the car price.

```python
fig, ax = plt.subplots()

ax.scatter(suv_df['mileage'], suv_df[ 'price'])
ax.set_title('Price and Mileage Comparison', fontweight='bold')
ax.set_xlabel('Mileage', fontweight='bold')
ax.set_ylabel('Price', fontweight='bold')

plt.show()
```

3. Check how many prices from the list are above a given threshold.

```python
PRICE_THRESHOLD = 45000

len(suv_df[suv_df['price'] > PRICE_THRESHOLD])
```

4. From the price list of the cars, compute the mean and standard deviation of the cost of an SUV. Print the results as a boxplot.

```python
cost_mean = suv_df['price'].mean()
cost_standard = suv_df['price'].std()

fig, ax = plt.subplots()

ax.boxplot([cost_mean, cost_standard])
ax.set_title("Mean and Standard Deviation of the Cost", fontweight='bold')
ax.set_ylabel('Price', fontweight='bold')

plt.show()
```

```python
fig, ax = plt.subplots()

ax.boxplot(suv_df['price'])
ax.set_title("Whole Price Column (2)", fontweight='bold')
ax.set_ylabel('Price', fontweight='bold')

plt.show()
```
