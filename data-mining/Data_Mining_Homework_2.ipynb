## 4. Implement an algorithm for incremental selection of attributes by using:

```python
import pandas as pd
import numpy as np

weather_play_data_set = pd.DataFrame(np.array([
  ['sunny', 'hot', 'high', False, False],
  ['sunny', 'hot', 'high', True, False],
  ['overcast', 'hot', 'high', False, True],
  ['rainy', 'mild', 'high', False, True],
  ['rainy', 'cool', 'normal', False, True],
  ['rainy', 'cool', 'normal', True, False],
  ['overcast', 'cool', 'normal', True, True],
  ['sunny', 'mild', 'high', False, False],
  ['sunny', 'cool', 'normal', False, True],
  ['rainy', 'mild', 'normal', False, True],
  ['sunny', 'mild', 'normal', True, True],
  ['overcast', 'mild', 'high', True, True],
  ['overcast', 'hot', 'normal', False, True],
  ['rainy', 'mild', 'high', True, False],
]), columns=['outlook', 'temperature', 'humidity', 'windy', 'play'])

weather_play_data_set
```

### a. Gini index

```python
from collections import Counter

CLASSES_COLUMN_NAME = 'play'

def get_gini(classes_column):
  nr_of_instances = len(classes_column)
  nr_of_instances_per_classes = Counter(classes_column).values()
  gini = 1

  for nr_of_instances_per_class in nr_of_instances_per_classes:
    probability = nr_of_instances_per_class / nr_of_instances

    gini -= probability ** 2

  return gini

def get_gini_for_attribute_value(
    data_set,
    attribute_name,
    attribute_value
):
  return get_gini(data_set.loc[
      data_set[attribute_name] == attribute_value
  ][CLASSES_COLUMN_NAME])

def get_gini_index_of_attribute(data_set, attribute_name):
  total_nr_of_instances = data_set.shape[0]
  attribute_values = list(
      Counter(data_set[attribute_name]).keys()
  )
  nr_of_instances_per_attributes_values = list(
      Counter(data_set[attribute_name]).values()
  )
  gini_index = 0

  for i in range(len(attribute_values)):
    gini_index_for_attribute_value = get_gini_for_attribute_value(
        data_set,
        attribute_name,
        attribute_values[i]
    )

    gini_index += nr_of_instances_per_attributes_values[i] \
      / total_nr_of_instances * gini_index_for_attribute_value

  return gini_index

def get_attributes_gini_index(data_set):
  gini_index_per_attributes = {}

  for column_name in data_set.columns:
    if (column_name != CLASSES_COLUMN_NAME):
      gini_index_for_attribute = get_gini_index_of_attribute(
          data_set,
          column_name
      )

      gini_index_per_attributes[
          column_name
      ] = gini_index_for_attribute

  return gini_index_per_attributes

attributes_gini_index = get_attributes_gini_index(
    weather_play_data_set
)

ATTRIBUTE_NAME_COLUMN_NAME = 'Attribute Name'
GINI_INDEX_COLUMN_NAME = 'Gini Index'

attributes_gini_index_df = pd.DataFrame(
    attributes_gini_index.items(),
    columns=[ATTRIBUTE_NAME_COLUMN_NAME, GINI_INDEX_COLUMN_NAME]
)

attributes_gini_index_df.sort_values(by=[GINI_INDEX_COLUMN_NAME])
```

### b. Information Gain (similar to Mutual Information)

```python
import math
from collections import Counter

CLASSES_COLUMN_NAME = 'play'

def get_entropy(classes_column):
  nr_of_instances = len(classes_column)
  nr_of_instances_per_classes = Counter(classes_column).values()
  partial_entropy = 0

  for nr_of_instances_per_class in nr_of_instances_per_classes:
    probability = nr_of_instances_per_class / nr_of_instances

    partial_entropy += probability * math.log2(probability)

  return -1 * partial_entropy

def get_entropy_of_attribute_value(
    data_set,
    attribute_name,
    attribute_value
):
  return get_entropy(data_set.loc[
      data_set[attribute_name] == attribute_value
  ][CLASSES_COLUMN_NAME])

def get_information_gain_of_attribute(
    data_set,
    attribute_name,
    classes_entropy
):
  total_nr_of_instances = data_set.shape[0]
  attribute_values = list(
      Counter(data_set[attribute_name]).keys()
  )
  nr_of_instances_per_attributes_values = list(
      Counter(data_set[attribute_name]).values()
  )
  partial_information_gain = 0

  for i in range(len(attribute_values)):
    entropy_for_attribute_value = get_entropy_of_attribute_value(
        data_set,
        attribute_name,
        attribute_values[i]
    )
    partial_information_gain += nr_of_instances_per_attributes_values[i] \
      / total_nr_of_instances * entropy_for_attribute_value

  return classes_entropy - partial_information_gain

def get_attributes_information_gains(data_set):
  information_gain_per_attributes = {}
  classes_entropy = get_entropy(data_set[CLASSES_COLUMN_NAME])

  for column_name in data_set.columns:
    if (column_name != CLASSES_COLUMN_NAME):
      information_gain_for_attribute = get_information_gain_of_attribute(
          data_set,
          column_name,
          classes_entropy
      )

      information_gain_per_attributes[
          column_name
      ] = information_gain_for_attribute

  return information_gain_per_attributes

attributes_information_gains = get_attributes_information_gains(
    weather_play_data_set
)

ATTRIBUTE_NAME_COLUMN_NAME = 'Attribute Name'
INFORMATION_GAIN_COLUMN_NAME = 'Information Gain'

attributes_information_gains_df = pd.DataFrame(
    attributes_information_gains.items(),
    columns=[ATTRIBUTE_NAME_COLUMN_NAME, INFORMATION_GAIN_COLUMN_NAME]
)

attributes_information_gains_df.sort_values(
    by=[INFORMATION_GAIN_COLUMN_NAME],
    ascending=False
)
```
