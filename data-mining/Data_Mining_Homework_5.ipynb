```python
from scipy.io import arff
import pandas as pd

supermarket_arff_data_set = arff.loadarff("/content/gdrive/MyDrive/DataMining/supermarket.arff")
supermarket_data_set = pd.DataFrame(supermarket_arff_data_set[0])

supermarket_data_set
```

## 1. Find association rules with a support of at least 0.2 and a confidence of at least 0.7 based on the list of transactions from supermarket.arff.

```python
supermarket_data_set = supermarket_data_set.iloc[:,~supermarket_data_set.columns.str.startswith('department')]
supermarket_data_set = supermarket_data_set.drop('total', axis=1)
supermarket_data_set = supermarket_data_set.select_dtypes([object])
supermarket_data_set = supermarket_data_set.stack().str.decode('utf-8').unstack()

supermarket_data_set
```

```python
transactions = []

for index, row in supermarket_data_set.iterrows():
    transaction = []

    for column_name in supermarket_data_set.columns:
        if row[column_name] == 't':
            transaction.append(column_name)

    transactions.append(transaction)

print(len(transactions))
```

```python
%pip install apyori
```

```python
from apyori import apriori

results = list(apriori(transactions))
selected_rules = list(apriori(transactions, min_support=0.2, min_confidence=0.7))

selected_rules
```
