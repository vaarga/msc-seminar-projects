## 1. Identify an appropriate regression model to estimate “miles per gallon” depending on various characteristics of the dataset (autoMPG.arff) and analyze the differences between models (particularly with respect to the regression trees).

```python
autoMpg_data_set = autoMpg_data_set.select_dtypes([object])
autoMpg_data_set = autoMpg_data_set.stack().str.decode('utf-8').unstack().astype(float)
autoMpg_data_set['displacement'] = autoMpg_data_set_original['displacement']
autoMpg_data_set['horsepower'] = autoMpg_data_set_original['horsepower']
autoMpg_data_set['weight'] = autoMpg_data_set_original['weight']
autoMpg_data_set['acceleration'] = autoMpg_data_set_original['acceleration']
autoMpg_data_set['class'] = autoMpg_data_set_original['class']
autoMpg_data_set
```

```python
autoMpg_data_set = autoMpg_data_set.dropna()
```

```python
from sklearn.model_selection import train_test_split

autoMpg_y = autoMpg_data_set['class']
autoMpg_X = autoMpg_data_set.drop(['class'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(
    autoMpg_X, autoMpg_y, test_size=0.3, random_state=42
)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score

def fit_predict_evaluate(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    mse = mean_squared_error(y_test, preds)
    r2 = r2_score(y_test, preds)

    return mse, r2

linear_regression = LinearRegression()
decision_tree = DecisionTreeRegressor()
random_forest = RandomForestRegressor()
svr = SVR()

linear_regression_mse, linear_regression_r2 = fit_predict_evaluate(linear_regression, X_train, y_train, X_test, y_test)
decision_tree_mse, decision_tree_r2 = fit_predict_evaluate(decision_tree, X_train, y_train, X_test, y_test)
random_forest_mse, random_forest_r2 = fit_predict_evaluate(random_forest, X_train, y_train, X_test, y_test)
svr_mse, svr_r2 = fit_predict_evaluate(svr, X_train, y_train, X_test, y_test)

print("Linear Regression MSE:", linear_regression_mse)
print("Decision Tree MSE:", decision_tree_mse)
print("Random Forest MSE:", random_forest_mse)
print("SVR MSE:", svr_mse)
print("-------------------------------------------")
print("Linear Regression R2 score:", linear_regression_r2)
print("Decision Tree R2 score:", decision_tree_r2)
print("Random Forest R2 score:", random_forest_r2)
print("SVR R2 score:", svr_r2)
```

An appropriate regression model to estimate “miles per gallon” depending on various characteristics of the dataset (autoMPG.arff) could be Random Forest (best MSE) or SVR (best R2 score).

## 2. Implement an RBF network for the estimation of “miles per gallon” starting from the example https://github.com/ameerezae/RBF-Network/blob/master/RBF.ipynb  and modifying:

### a) the random selection of the centers is replaced with a selection based on clustering

### b) the parameter corresponding to the width of the receptive field (sigma)  is computed using the positions of the centers – see examples of heuristics in Lecture 10)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

class RBF(object):
    def __init__(self, shape):
        self.shape = shape
        self.sigma = 0
        self.centers = None
        self.weights = None

    def gaussian_kernel(self, center, data_point):
        return np.exp(-1 * self.sigma * (np.linalg.norm(center - data_point) ** 2))

    def interpolation_matrix(self, X):
        matrix = np.zeros(len(X) * self.shape)
        counter = 0

        for inp in X:
            for center in self.centers:
                kernel = self.gaussian_kernel(center, inp)
                matrix[counter] = kernel
                counter += 1

        return matrix.reshape(len(X), self.shape)

    def select_centers(self, X):
        kmeans = KMeans(n_clusters=self.shape, random_state=0)
        kmeans.fit(X)
        centers = kmeans.cluster_centers_

        return centers

    def compute_sigma(self, centers):
        max_distance = np.max(np.linalg.norm(centers[:, np.newaxis] - centers, axis=-1))
        sigma = max_distance / np.sqrt(2 * self.shape)

        return sigma

    def fit(self, X, Y):
        self.centers = self.select_centers(X)
        self.sigma = self.compute_sigma(self.centers)
        im = self.interpolation_matrix(X)
        self.weights = np.dot(np.linalg.pinv(im), Y)

    def predict(self, X):
        im = self.interpolation_matrix(X)
        predictions = np.dot(im, self.weights)

        return predictions

x_train_RBF = autoMpg_X.to_numpy()
y_train_RBF = autoMpg_y.to_numpy()

RBFN = RBF(shape=10)
RBFN.fit(x_train_RBF, y_train_RBF)
y_pred_RBF = RBFN.predict(x_train_RBF)
```
